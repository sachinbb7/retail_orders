```python
#create discount to understand the discount value
df['discount']= (df['discount_percent'] * df['list_price'])/100

#create sales price
df['sales_price']= df['list_price']- df['discount']

#create profit
df['profit']= df['sales_price']- df['cost_price']
```

```python
df['order_date']= pd.to_datetime(df['order_date'], format ="%Y-%m-%d")
```

```python
df.drop(columns= ['discount_percent','cost_price', 'list_price'], inplace =True)
```

```python
# Ensure order_date is datetime
df["order_date"] = pd.to_datetime(df["order_date"])

# Extract month-year for grouping
df["month_year"] = df["order_date"].dt.to_period("M")

# Compute mode of ship_mode for each month
mode_map = df.groupby("month_year")["ship_mode"].agg(lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan)

# Fill NaNs with mode of that month
df["ship_mode"] = df.apply(
    lambda row: mode_map[row["month_year"]] if pd.isna(row["ship_mode"]) else row["ship_mode"],
    axis=1
)

# Drop helper column if you want
df.drop(columns="month_year", inplace=True)
```
